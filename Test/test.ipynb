```python
import requests
from bs4 import BeautifulSoup

# Replace this with the actual page URL you’re scraping:
url = "https://hindi-kavita.com/Hindi-Ramdhari-Singh-Dinkar.php"

# 1. Fetch the page
response = requests.get(url)
html_content = response.text

# 2. Parse the HTML
soup = BeautifulSoup(html_content, "html.parser")

# 3. Identify the links by a class or HTML structure
#    (In your screenshot, each poem link appears to have class="poem")
links = soup.select("a")

# 4. Print out the link text and href
for link in links:
    poem_title = link.get_text(strip=True)
    poem_url = link.get("href")
    print(f"Title: {poem_title}")
    print(f"Link:  {poem_url}")
    print("-" * 40)


# save all poem url in a list
poem_url_list = []
for link in links:
    poem_url = link.get("href")
    poem_url_list.append(poem_url)
```

```python
def scrape_poems_from_url(url):
    try:
        # Add headers to mimic a browser request
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
        }
        
        # Fetch and parse the page
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise an exception for bad status codes
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Get all poem sections
        poem_sections = soup.select("div.section div.list h3, div.section div.list p")
        
        if not poem_sections:
            print(f"Warning: No poem sections found in {url}")
            return []
            
        # Initialize variables
        poems_data = []
        current_title = ""
        
        # Extract poems with their titles
        for element in poem_sections:
            if element.name == 'h3':
                try:
                    # Handle cases where there might not be a period in the title
                    if '.' in element.text:
                        current_title = element.text.split('.', 1)[1].strip()
                    else:
                        current_title = element.text.strip()
                except Exception as e:
                    print(f"Warning: Error processing title in {url}: {str(e)}")
                    current_title = element.text.strip()
            elif element.name == 'p':
                try:
                    poem_text = element.get_text('\n').strip()
                    if current_title and poem_text:
                        poems_data.append(f"Source: {url}\nTitle: {current_title}\n\n{poem_text}\n\n{'='*50}\n")
                except Exception as e:
                    print(f"Warning: Error processing poem text in {url}: {str(e)}")
        
        return poems_data
        
    except requests.exceptions.RequestException as e:
        print(f"Error: Failed to fetch {url}: {str(e)}")
        return []
    except Exception as e:
        print(f"Error: Unexpected error processing {url}: {str(e)}")
        return []
    
# Process all URLs and save to file
successful_count = 0
failed_count = 0

with open('data.txt', 'w', encoding='utf-8') as f:
    for url in poem_url_list:
        print(f"Processing: {url}")
        poems = scrape_poems_from_url(url)
        
        if poems:
            for poem in poems:
                f.write(poem)
            successful_count += 1
        else:
            failed_count += 1
            
        # Add delay between requests
        time.sleep(2)

print(f"\nScraping completed:")
print(f"Successfully processed: {successful_count} URLs")
print(f"Failed to process: {failed_count} URLs")
print("All poems have been saved to data.txt")
```

```python
def format_poem_text(poem_text):
    # Split the poem into lines
    lines = [line.strip() for line in poem_text.split('\n') if line.strip()]
    
    # Group lines into four-line stanzas
    formatted_stanzas = []
    current_stanza = []
    
    for line in lines:
        current_stanza.append(line)
        if len(current_stanza) == 4:
            formatted_stanzas.append('\n'.join(current_stanza))
            current_stanza = []
    
    # Add any remaining lines
    if current_stanza:
        formatted_stanzas.append('\n'.join(current_stanza))
    
    # Join stanzas with double newline
    return '\n\n'.join(formatted_stanzas)

def scrape_poems_from_url(url):
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
        }
        
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")
        
        poem_sections = soup.select("div.section div.list h3, div.section div.list p")
        
        if not poem_sections:
            print(f"Warning: No poem sections found in {url}")
            return []
            
        poems_data = []
        current_title = ""
        
        for element in poem_sections:
            if element.name == 'h3':
                try:
                    current_title = element.text.split('.', 1)[1].strip() if '.' in element.text else element.text.strip()
                except Exception as e:
                    print(f"Warning: Error processing title in {url}: {str(e)}")
                    current_title = element.text.strip()
            elif element.name == 'p':
                try:
                    poem_text = element.get_text('\n').strip()
                    if current_title and poem_text:
                        formatted_poem = format_poem_text(poem_text)
                        poems_data.append(f"Source: {url}\nTitle: {current_title}\n\n{formatted_poem}\n\n{'='*50}\n")
                except Exception as e:
                    print(f"Warning: Error processing poem text in {url}: {str(e)}")
        
        return poems_data
        
    except requests.exceptions.RequestException as e:
        print(f"Error: Failed to fetch {url}: {str(e)}")
        return []
    except Exception as e:
        print(f"Error: Unexpected error processing {url}: {str(e)}")
        return []

# Process all URLs and save to file
successful_count = 0
failed_count = 0

with open('data1.txt', 'w', encoding='utf-8') as f:
    for url in poem_url_list:
        print(f"Processing: {url}")
        poems = scrape_poems_from_url(url)
        
        if poems:
            for poem in poems:
                f.write(poem)
            successful_count += 1
        else:
            failed_count += 1
            
        time.sleep(2)

print(f"\nScraping completed:")
print(f"Successfully processed: {successful_count} URLs")
print(f"Failed to process: {failed_count} URLs")
print("All poems have been saved to data.txt")
```

```python
def clean_file(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as f:
        lines = f.readlines()
    
    # Filter out unwanted lines
    cleaned_lines = [
        line for line in lines 
        if not line.startswith('======') 
        and not line.startswith('Source:')
    ]
    
    # Write cleaned content to new file
    with open(output_file, 'w', encoding='utf-8') as f:
        f.writelines(cleaned_lines)

# Clean the file
clean_file('data1.txt', 'cleaned_data.txt')

# Verify the results
with open('cleaned_data.txt', 'r', encoding='utf-8') as f:
    print("First few lines of cleaned file:")
    print(f.read(200))
```
